```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
import sys
sys.path.append(rf"/Users/baia/Desktop/PYTHON/mba_dsa_usp_esalq")

from TCC.utils.constantes import *
```

```python
# 1. Defina a lista com TODAS as variáveis macroeconômicas que você quer testar juntas
# (Não inclua o Bitcoin aqui, pois o VIF avalia apenas as variáveis independentes X)
variaveis_X = [
    'spx_log_ret', 
    'nasdaq_log_ret', 
    'dxy_log_ret', 
    'vix_log_ret', 
    'gold_log_ret', 
    'us10y_diff'
    # Adicione outras do seu bloco macro, se houver
]

# 2. Separe o DataFrame apenas com essas variáveis
# (Assumindo que sua base principal se chama df)
X = df_features_transformada[variaveis_X]

# 3. PASSO CRÍTICO: Adicionar a constante (intercepto)
X_com_constante = sm.add_constant(X)

# 4. Criar um DataFrame vazio para armazenar os resultados formatados
vif_data = pd.DataFrame()
vif_data["Variável"] = X_com_constante.columns

# 5. Calcular o VIF para cada variável usando um list comprehension
vif_data["VIF"] = [variance_inflation_factor(X_com_constante.values, i) 
                   for i in range(X_com_constante.shape[1])]

# 6. Exibir o resultado arredondado, removendo a linha da 'const' que não nos interessa
vif_data = vif_data[vif_data['Variável'] != 'const'].round(2)
vif_data = vif_data.sort_values(by="VIF", ascending=False).reset_index(drop=True)

print("=== DIAGNÓSTICO DE MULTICOLINEARIDADE (VIF) ===")
print(vif_data)
```
